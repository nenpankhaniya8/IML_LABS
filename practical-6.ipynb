# to convert a NumPy array to a Pandas series

```python
import numpy as np
import pandas as pd

ser = pd.Series()
print("Pandas Series: ", ser)

data = np.array(['m', 'a', 'h', 'e', 'k'])
ser = pd.Series(data)
print("Pandas Series:\n", ser)
```

# to convert the first column of a DataFrame as a Series

```python
import pandas as pd

df = pd.DataFrame()
print(df)

lst = ['Geeks', 'For', 'Geeks', 'is', 'portal', 'for', 'Geeks']
df = pd.DataFrame(lst)
print("Dataframe")
print(df)
```

# to create the mean and standard deviation of the data of a given Series 

```python
import pandas as pd
s = pd.Series(data = [9, 9, 8, 5, 7, 8, 1, 2, 3, 4, 5, 6, 7, 8, 9, 5, 3])
print(s.mean())
print(s.std())
```

# to sort a given Series

```python
import pandas as pd
sr = pd.Series([11.6, 15.1, 22.18, 29.134, 17.1102])
print(sr)
print(sr.sort_values(ascending = True))
```
